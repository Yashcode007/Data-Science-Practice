```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# pie graphs 
labels=["A","B","C","D"]
sizes = [3,4,6,2]
colors=["blue", "purple", "red", "pink"]
explode=[0.1,0.2,0,0]
plt.title("Split among classes")
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=100)
plt.axis("equal")
plt.pie(sizes)
plt.show()
```
