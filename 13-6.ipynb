```python
import matplotlib.pyplot as plt
```

```python
year= [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
plt.scatter(year,salary, s=population)
plt.show()
```

```python
plt.scatter(year,salary,s=population,c='r',alpha=0.75,edgecolor='black')
plt.title("Bubble Chart ")
plt.show()
```

```python
plt.scatter(year,salary,s=population,alpha=0.5,edgecolor='black',marker='^')
plt.title("Bubble Chart")
plt.text(year[0]+0.1 , salary[0]+0.1,population[0])
plt.text(year[4]+0.1 , salary[4]+0.1,population[4])
plt.show()
```
